## Liste Kavramı

```python
import pandas as pd
sozluk = {"isim": ["ali", "veli", "kenan", "murat", "ayse", "hilal"],
          "yas": [15,16,17,33,45,66],
          "sehir": ["izmir", "Ankara", "Konya", "Ankara", "Ankara", "Antalya"]
}
veri = pd.DataFrame(sozluk)
veri
```

```python
# ortalama yasi bulalım
ortalama_yas = veri.yas.mean()
#np.mean(veri.yas) yöntemini de kullanabilirdik
ortalama_yas
```

```python
veri["YAS_GRUBU"] = ["kucuk" if ortalama_yas > i else "buyuk" for i in veri.yas]
veri
```

```python
veri.columns = [i.lower() for i in veri.columns] #kolon başlıkları küçültür
veri
```

```python
# assign fonksiyonu atama yapar. veri.assign 
veri = veri.assign(yas_5_yil_sonra = lambda x: (x['yas'] + 5))
veri
```

```python
def yas10yilsonra(age):
    return age+10
```

```python
veri["yas_10_yil_sonra"] = veri.yas.apply(yas10yilsonra)
```

```python
veri
```

## Birleştirme

```python
import pandas as pd
sozluk1 = { "isim": ["ali", "veli", "kenan"],
           "yas" : [15, 16, 17],
           "sehir" : ["izmir","Ankara","Konya"]}
veri1 = pd.DataFrame(sozluk1)
veri1
```

```python
# dikey birleştirme; axis = 0 ise dikey
veri_dikey = pd.concat([veri, veri1], axis = 0) ## concat artarda ekle ya da uçuca ekle
veri_dikey
```

```python
# yatay birleştirme ; axis = 1 ise yatay
veri_yatay = pd.concat([veri,veri1], axis=1)
veri_yatay
```
