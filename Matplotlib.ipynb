## Saçılım Grafiği

```python
import matplotlib.pyplot as plt
import pandas as pd

veri = pd.read_csv("iris.csv")
veri.head()
```

```python
# şimdi sepal length sütununu id'ye göre çizelim
plt.figure()
plt.scatter(veri["SepalLengthCm"], veri["SepalWidthCm"], color = "blue", s = 15, alpha = 0.7) # alpha saydamlık, a : size
plt.title("Alt Yaprak Uzunluğu ve Genişliği")
plt.xlabel("Alt Yaprak Uzunluğu (Cm)")
plt.ylabel("Alt Yaprak Genişliği (Cm)")
plt.grid(True)
plt.show
```

## Hisftogram Grafiği

```python
import matplotlib.pyplot as plt
import pandas as pd
veri = pd.read_csv("iris.csv")
veri.head()
```

```python
# şimdi sepal length sutununu id'ye göre çizdirelim
plt.figure()
plt.hist(veri["SepalLengthCm"], color = "green", alpha =0.5, bins = 20, label = "Alt Yaprak Uzunluğu") # bins : aralık demek. örnekte 20 girildi yani 20 eşit parçaya bölünüyor.
plt.hist(veri["PetalLengthCm"], color = "orange", alpha = 0.5, bins = 20, label= "Üst Yaprak Uzunluğu")
plt.title("Alt Yaprak Uzunluğu ve Üst Yaprak Uzunluğu - Histogram")
plt.xlabel("Cm")
plt.ylabel("Frekans")
plt.legend() # label ' ların görünebilmesi için
```

## Çubuk Grafiği

```python
import matplotlib.pyplot as plt
import pandas as pd

veri = pd.read_csv("iris.csv")
veri.head()
```

```python
veri_sutun_listesi = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
tanim = veri.describe() 
tanim
```

```python
# alt yaprak uzunluk ve genişlik , üst yaprak uzunluk ve genişlik ortalama değerlerini bulalim.
ortalama = tanim.iloc[1,[1,2,3,4]] # mean() satırını alıyoruz. 1 atırın 1,2,3,4 sutunlarını al.
ortalama
```

```python
# alt-üst yaprak uzunluk ve genişlik değerlerinin ortalamalarını bar plot ile çizdirelim.
plt.figure()
plt.bar(veri_sutun_listesi, ortalama)
plt.title('Alt-Üst Yaprak Uzunluğu ve Genişliği Ortalama')
plt.ylabel("Cm")
plt.show()
```

## Alt Plan Grafiği

```python
import matplotlib.pyplot as plt
import pandas as pd
veri = pd.read_csv("iris.csv")
veri.head()
```

```python
fig, ax = plt.subplots(2, 1) # figure üzerine 2*1 lik bir tablo oluşturur. bu tablo üzerine ax( eksen) ekle
fig.suptitle('Alt Yaprak Uzunluğu ve Üst Yaprak Uzunluk', fontsize= 14)
ax[0].plot(veri.Id, veri.SepalLengthCm, color = "red", label = "SepalLengthCm")
ax[0].set_title('Alt Yaprak Uzunluk') # başlık
ax[0].set_xlabel('id') # x ekseni label
ax[0].set_ylabel('Cm') # y ekseni label

ax[1].plot(veri.Id, veri.PetalLengthCm, color = 'green', label= 'PetalLengthCm')
ax[1].set_title('ÜSt Yaprak Uzunluk')
ax[1].set_xlabel('id')
ax[1].set_ylabel('Cm')

fig.tight_layout() # figure arası yerleşimi ayarla
fig.subplots_adjust(top=0.85) # en üstten ayarlanması sağlanır
plt.show()
```
