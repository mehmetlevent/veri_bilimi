# Gereksiz verinin çıkarılması ve düzenlenmesi

```python
# drop metadu ile id ve oyunları çıkaralım
veri = veri.drop(["id", "oyunlar"], axis = 1) # axis 1 sütunları çıkarmak için
veri.head()
```

```python
# NaN yani not a number . veri yok ya da kayıp. bunlarıdüzeltelim. boy ve kilo verilerini etkinlik ortalamalarına göre dolduralım.
essiz_etkinlik = np.unique(veri.etkinlik)
print("essiz etkinlik sayisi : {}".format(len(essiz_etkinlik)))
essiz_etkinlik[:10]
```

```python
# her bir etkinliği iteratif olarak dolaş
#etkinlik özelinde boy ve kilo ortalamalarını bul
# etkinlik özelinde boy ve kilo da kayıp olan değerlere ortalama boy ve kilo değerlerini eşitle
veri_gecici = veri.copy() # gerçek veriyi kaybetmemek için kopyalıyoruz.
boy_kilo_liste = ["boy", "kilo"]
for e in essiz_etkinlik: # etkinlik içerisinde dolaş
    # etkinlik filtresi oluştur
    etkinlik_filtre = veri_gecici.etkinlik == e
    # veriyi etkinliğe göre filtrele
    veri_filtreli = veri_gecici[etkinlik_filtre]
    # boy ve kilo özelinde ortalama hesapla
    for s in boy_kilo_liste:
        ortalama = np.round(np.mean(veri_filtreli[s]),2)
        if ~np.isnan(ortalama): # eğer etkinlik özelinde ortalama hesaplanabiliyorsa; isnan geçirli sayı değilmi diye bakıyor. ~ise değildir
            veri_filtreli[s] = veri_filtreli[s].fillna(ortalama)
        else: # etkinlik özelinde bir ortalama hesaplanamıyorsa tüm veri için ortalama hesapla
            tum_veri_ortalamasi = np.round(np.mean(veri[s]),2)
            veri_filtreli[s] = veri_filtreli[s].fillna(tum_veri_ortalamasi)
            # etkinlik özelinde kayıp değerleri doldurulmuş olan veriyi veri_gecic değişkenine eşitle
        veri_gecici[etkinlik_filtre] = veri_filtreli
    # kayıp değerleri giderilmiş çeçici veriyi gerçek veri değişkenine eşitle
veri = veri_gecici.copy()
veri.info() # boy ve kilo üzerlerinde kayıp değer olup olmadığını kontrol ediyoruz.
```

```python
# yaş sütununda bulunana kayıp veriyi, veri setinin ortalamasına göre dolduracağız
# yaş değişkeni tanımlı olmayan verileri bul
# tilda işareti ile tersini al
# yaş değişkeni tanımlı olan verileri bulmak için filtre oluştur
yas_ortalamasi = np.round(np.mean(veri.yas),2)
print("Yas ortalamasi: {}".format(yas_ortalamasi))
veri["yas"] = veri["yas"].fillna(yas_ortalamasi)
veri.info()
```

```python
# öncelikli olarak bar grafiğini elde edeceğimiz metodumuzu yazalım
def plotBar(degisken, n=5):
    """
        girdi: değişken/sutun ismi
            n= en önemli 5 eşsiz değer
        çıktı: Bar grafiği
    """
    veri_ = veri[degisken]
    veri_sayma = veri_.value_counts()
    veri_sayma = veri_sayma[:n]
    plt.figure()
    plt.bar(veri_sayma.index, veri_sayma, color ="orange")
    plt.xticks(veri_sayma.index, veri_sayma.index.values)
    plt.xticks(rotation = 45) # 45 derecelik açıyla yaz
    plt.ylabel("Frekans")
    plt.title("Veri Sıklığı - {}".format(degisken))
    plt.show()
    print("{} : \n{}".format(degisken, veri_sayma))
    
```

## anomali testi

```python
# anamoli tespiti için fonksiyon yazalım
def anamoliTespiti(df,ozellik):
    outlier_indices = []
    
    for c in ozellik:
        # 1. çeyrek
        Q1 = np. percentile(df[c],25) # percentile yüzdelik dilim demek
        # 3. çeyrek
        Q3 = np.percentile(df[c],75)
        # IQR : Çeyrekler açıklığı
        IQR = Q3 - Q1
        # aykırı tespiti için çarpan değeri
        outlier_step = IQR * 1.5
        outlier_list_col = df[(df[c] < Q1 - outlier_step) | (df[c] > Q3 + outlier_step ) ].index
        # indeksleri depola
        outlier_indices.extend(outlier_list_col)
    # eşsiz aykırı değerleri bul
    outlier_indices = Counter(outlier_indices)
    # eğer bir örnek 1 farklı sütun için aykırı değerse bunu aykırı değer olarak kabul et
    multiple_outliers = list(i for i, v in outlier_indices.items() if v >1 )
    
    return multiple_outliers
```

```python
veri_anomali = veri.loc[anamoliTespiti(veri, ["yas", "kilo", "boy"])]
veri_anomali.spor.value_counts()
```

```python
# veri içerisinde bulunan yılları datetime veri tipine dönüştürelim
tarih_saat_nesnesi = pd.to_datetime(veri_zaman['yil'], format="%Y")
print(type(tarih_saat_nesnesi))
tarih_saat_nesnesi.head(3)
```

```python
# veriye tarih  saat diye yeni bir sütun açıp, bir önceki satırda elde ettiğimiz datetime veri tipini ekleyelim
veri_zaman["tarih_saat"] = tarih_saat_nesnesi
veri_zaman.head(3)
```

```python
# tarih_saat sütununda bulunan datetime veri tipine ait veriyi, asıl verinin indeksi yapalım
# pandas kütüphanesinde indeksi datetime veri tipi olan veri setleri ile çalşmak için özel yapılar bulunmaktadır.
# bu nedenle amacımız olan indeksi datetime veri tipi yapma çalışmamız gerçeklerşmiş oluyor.
veri_zaman= veri_zaman.set_index("tarih_saat")
veri_zaman.drop(["yil"], axis = 1, inplace = True)
veri_zaman.head(3)
```

```python
periyodik_veri = veri_zaman.resample("2A").mean() # 2 yıllık periyotlar halinde ortalama değerleri al
periyodik_veri.head()
```

```python
# kayıp verileri çıkaralım. NaN
periyodik_veri.dropna(axis=0,inplace =True)  # axis= 0 :satır; dropna : anlamsız verileri kaldır NaN
periyodik_veri.head()
```

```python
plt.figure()
periyodik_veri.plot()
plt.title("Yıllara göre yaş, boy ve kilo değişimi")
plt.xlabel("yıl")
plt.grid(True)
plt.show()
```
