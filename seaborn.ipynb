```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
veri = pd.read_csv("olimpiyatlar_temizlenmis.csv")
veri.head()
```

```python
# seaborn ile salm grafiini izelim
def sacilimGrafik():
    sns.scatterplot(x = "boy", y = "kilo", data = veri)
    plt.title('boy ve kilo dalm - beyaz zgara Tema')
    plt.show()
sacilimGrafik()
```

```python
# seaborn stilleri
sns.set_style("white")
sacilimGrafik()
```

```python
sns.set_style("darkgrid")
sacilimGrafik()
```

```python
sns.set_style("whitegrid")
sacilimGrafik()
```

```python
# salm grafii -kategorik
# hue = snf ya da etiket olarak adlandrlabilir
# scatter plot izdirilirken hue ' ya gre snflar farkl renklendirilir
sns.scatterplot(x = "boy", y = "kilo", hue = "madalya", data = veri) # hue = renk nuans anlamnda
plt.title("madalya tipine gre boy ve kilo dalm")
plt.show()
```

```python
# salm grafii- dorusal regresyon
# salm grafii izilir, zerine dorusal regresyon yntemi ile line fit ilemi gerekltirilir.
# dorusal regresyon kavram, makine renmesi konusu ierisinde anlatlacaktr.
sns.regplot(x="boy", y ="kilo", data =veri, marker ="+", scatter_kws= {'alpha':0.2})# regresyonun reg i
plt.title("boy ve kilo salm")
plt.show()
```

```python
# REnk paletleri
sns.scatterplot(x="boy", y="kilo", hue ="madalya", data=veri, palette ="Set1")
plt.title("Farkl madalya trlerine gre boy ve kilo dalm")
plt.show()
```

```python
sns.scatterplot(x="boy", y="kilo", data =veri, hue = "madalya", palette="rocket") # scatterplot seaborn da salm grafiini izer
plt.title("boy kilo dalm")
plt.show()
```

```python
# izgi grafii :lineplot
sns.lineplot(x="boy", y="kilo", data = veri)
plt.title("boy ve kilo")
plt.show()
```

```python
# izgi grafii - kategorik
sns.lineplot(x="boy", y="kilo",  hue="cinsiyet", data = veri)
plt.title("cinsiyete gre boy ve kilo bamllklar")
plt.show()
```

```python
# histogram : displot
sns.displot(veri, x ="kilo")
plt.ylabel("Frekans")
plt.title("kilo histogram")
plt.show()
```

```python
# histogram - kategorik
sns.displot(veri, x="kilo", hue ="cinsiyet")
plt.ylabel("Frekans")
plt.title("cinsiyete gre histogram")
plt.show()
```

```python
sns.displot(veri, x="kilo", col="cinsiyet", multiple="dodge") # multiple="dodge") iki ayr histogram izdirir
plt.title("kilo")
plt.show()
```

```python
# iki boyutlu histogram
sns.displot(veri, x="kilo", y ="boy", kind="kde") # kind :tr
plt.xlabel("kilo")
plt.ylabel("boy")
plt.title("kilo ve boy histogram")
plt.show()
```

```python
sns.displot(veri, x="kilo", y="boy", hue ="cinsiyet", kind="kde")
plt.ylabel("boy")
plt.xlabel("kilo")
plt.title("boy kilo cinsiyete gre 2 boyutlu histogram")
plt.show()
```

```python
sns.kdeplot(data=veri, x="kilo", y="boy", hue="cinsiyet",multiple="stack", fill=True)
plt.ylabel("boy")
plt.xlabel("kilo")
plt.title("boy kilo cinsiyete gre 2 boyutlu histogram")
plt.show()
```

```python
# seaborn ubuk grafii
sns.barplot(x="madalya", y="boy", data=veri)
plt.title("madalyalara gre boy grafii")
plt.show()
```

```python
sns.barplot(x="madalya", y="boy", hue="cinsiyet", data=veri)
plt.show()
```

```python
# madalya alm sporcular cinsiyete gre ayrtrmak
sns.catplot(x="madalya", y="yas", hue="cinsiyet", col="sezon", data=veri, kind="bar", height=4, aspect=.7)
plt.show()
```

```python
# seaborn kutu grafii
sns.boxplot(x="sezon", y="boy", data=veri)
plt.show()
```

```python
sns.boxplot(x="sezon", y="boy", hue ="cinsiyet", data=veri, palette ="Set2")
plt.show()
```

```python
veri_gecici= veri.loc[:,["yas","boy","kilo"]]
sns.boxplot(data=veri_gecici, orient ="h", palette ="Set2") # orint="h" : yatay olarak iz
plt.show()
```

```python
sns.catplot(x="sezon", y="boy", hue="cinsiyet", col="madalya", data=veri, kind="box", height=4, aspect=.7)
plt.show()
```

```python
# s haritas
sns.heatmap(veri.corr(), annot=True, linewidth=.5, fmt=".1f")
plt.show()
```

```python
# seaborn keman grafii
sns.violinplot(x="sezon", y="boy", data=veri)
plt.show()
```

```python
sns.violinplot(x="sezon", y="boy", hue="cinsiyet", data=veri)
plt.show()
```

```python
sns.violinplot(x="sezon", y="boy", data=veri, hue="cinsiyet", split=True)
plt.show()
```

```python
sns.catplot(x="sezon", y="boy", hue="cinsiyet", col="madalya", 
            data=veri, kind="violin", split=True, height =4, aspect=.7)
plt.show()
```

```python
#seaborn ortaklk grafii
sns.jointplot( data=veri, x="kilo", y="boy", hue="sezon", kind="kde")
plt.show()
```

```python
g = sns.JointGrid(data=veri, x="kilo", y="boy",hue="cinsiyet")
g.plot_joint(sns.histplot)
g.plot_marginals(sns.boxplot)
plt.show()
```

```python
#seaborn ift-e grafii
sns.pairplot(veri)  #4 satr 4 stun ikierli 
plt.show()
```

```python
g= sns.PairGrid(veri)
g.map_upper(sns.histplot)
g.map_lower(sns.kdeplot, fill=True)
g.map_diag(sns.histplot, kde=True)
plt.show()
```

```python
#seaborn sayma grafii
sns.countplot(x="sehir", data=veri)
plt.xticks(rotation= 90) # ehirlerin ismini 90 derece ile yaz
plt.show()
```
