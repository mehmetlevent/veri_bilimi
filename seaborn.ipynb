```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
veri = pd.read_csv("olimpiyatlar_temizlenmis.csv")
veri.head()
```

```python
# seaborn ile saçılım grafiğini çizelim
def sacilimGrafik():
    sns.scatterplot(x = "boy", y = "kilo", data = veri)
    plt.title('boy ve kilo dağılımı - beyaz ızgara Tema')
    plt.show()
sacilimGrafik()
```

```python
# seaborn stilleri
sns.set_style("white")
sacilimGrafik()
```

```python
sns.set_style("darkgrid")
sacilimGrafik()
```

```python
sns.set_style("whitegrid")
sacilimGrafik()
```

```python
# saçılım grafiği -kategorik
# hue = sınıf ya da etiket olarak adlandırılabilir
# scatter plot çizdirilirken hue ' ya göre sınıflar farklı renklendirilir
sns.scatterplot(x = "boy", y = "kilo", hue = "madalya", data = veri) # hue = renk nuans anlamında
plt.title("madalya tipine göre boy ve kilo dağılımı")
plt.show()
```

```python
# saçılım grafiği- doğrusal regresyon
# saçılım grafiği çizilir, üzerine doğrusal regresyon yöntemi ile line fit işlemi gerçeklştirilir.
# doğrusal regresyon kavramı, makine öğrenmesi konusu içerisinde anlatılacaktır.
sns.regplot(x="boy", y ="kilo", data =veri, marker ="+", scatter_kws= {'alpha':0.2})# regresyonun reg i
plt.title("boy ve kilo sağılımı")
plt.show()
```

```python
# REnk paletleri
sns.scatterplot(x="boy", y="kilo", hue ="madalya", data=veri, palette ="Set1")
plt.title("Farklı madalya türlerine göre boy ve kilo dağılımı")
plt.show()
```

```python
sns.scatterplot(x="boy", y="kilo", data =veri, hue = "madalya", palette="rocket") # scatterplot seaborn da saçılım grafiğini çizer
plt.title("boy kilo dağılımı")
plt.show()
```

```python
# çizgi grafiği :lineplot
sns.lineplot(x="boy", y="kilo", data = veri)
plt.title("boy ve kilo")
plt.show()
```

```python
# çizgi grafiği - kategorik
sns.lineplot(x="boy", y="kilo",  hue="cinsiyet", data = veri)
plt.title("cinsiyete göre boy ve kilo bağımlılıkları")
plt.show()
```

```python
# histogram : displot
sns.displot(veri, x ="kilo")
plt.ylabel("Frekans")
plt.title("kilo histogramı")
plt.show()
```

```python
# histogram - kategorik
sns.displot(veri, x="kilo", hue ="cinsiyet")
plt.ylabel("Frekans")
plt.title("cinsiyete göre histogram")
plt.show()
```

```python
sns.displot(veri, x="kilo", col="cinsiyet", multiple="dodge") # multiple="dodge") iki ayrı histogram çizdirir
plt.title("kilo")
plt.show()
```

```python
# iki boyutlu histogram
sns.displot(veri, x="kilo", y ="boy", kind="kde") # kind :tür
plt.xlabel("kilo")
plt.ylabel("boy")
plt.title("kilo ve boy histogram")
plt.show()
```

```python
sns.displot(veri, x="kilo", y="boy", hue ="cinsiyet", kind="kde")
plt.ylabel("boy")
plt.xlabel("kilo")
plt.title("boy kilo cinsiyete göre 2 boyutlu histogram")
plt.show()
```

```python
sns.kdeplot(data=veri, x="kilo", y="boy", hue="cinsiyet",multiple="stack", fill=True)
plt.ylabel("boy")
plt.xlabel("kilo")
plt.title("boy kilo cinsiyete göre 2 boyutlu histogram")
plt.show()
```

```python
# seaborn çubuk grafiği
sns.barplot(x="madalya", y="boy", data=veri)
plt.title("madalyalara göre boy grafiği")
plt.show()
```

```python
sns.barplot(x="madalya", y="boy", hue="cinsiyet", data=veri)
plt.show()
```

```python
# madalya almış sporcuları cinsiyete göre ayrıştırmak
sns.catplot(x="madalya", y="yas", hue="cinsiyet", col="sezon", data=veri, kind="bar", height=4, aspect=.7)
plt.show()
```

```python
# seaborn kutu grafiği
sns.boxplot(x="sezon", y="boy", data=veri)
plt.show()
```

```python
sns.boxplot(x="sezon", y="boy", hue ="cinsiyet", data=veri, palette ="Set2")
plt.show()
```

```python
veri_gecici= veri.loc[:,["yas","boy","kilo"]]
sns.boxplot(data=veri_gecici, orient ="h", palette ="Set2") # orint="h" : yatay olarak çiz
plt.show()
```

```python
sns.catplot(x="sezon", y="boy", hue="cinsiyet", col="madalya", data=veri, kind="box", height=4, aspect=.7)
plt.show()
```

```python
# ısı haritası
sns.heatmap(veri.corr(), annot=True, linewidth=.5, fmt=".1f")
plt.show()
```

```python
# seaborn keman grafiği
sns.violinplot(x="sezon", y="boy", data=veri)
plt.show()
```

```python
sns.violinplot(x="sezon", y="boy", hue="cinsiyet", data=veri)
plt.show()
```

```python
sns.violinplot(x="sezon", y="boy", data=veri, hue="cinsiyet", split=True)
plt.show()
```

```python
sns.catplot(x="sezon", y="boy", hue="cinsiyet", col="madalya", 
            data=veri, kind="violin", split=True, height =4, aspect=.7)
plt.show()
```

```python
#seaborn ortaklık grafiği
sns.jointplot( data=veri, x="kilo", y="boy", hue="sezon", kind="kde")
plt.show()
```

```python
g = sns.JointGrid(data=veri, x="kilo", y="boy",hue="cinsiyet")
g.plot_joint(sns.histplot)
g.plot_marginals(sns.boxplot)
plt.show()
```

```python
#seaborn çift-eş grafiği
sns.pairplot(veri)  #4 satır 4 sütun ikişerli 
plt.show()
```

```python
g= sns.PairGrid(veri)
g.map_upper(sns.histplot)
g.map_lower(sns.kdeplot, fill=True)
g.map_diag(sns.histplot, kde=True)
plt.show()
```

```python
#seaborn sayma grafiği
sns.countplot(x="sehir", data=veri)
plt.xticks(rotation= 90) # şehirlerin ismini 90 derece ile yaz
plt.show()
```
